```python
# For performing mathematical calculation
import numpy as np 
# For handling Dataframe or data
import pandas as pd
# For visualisation
import matplotlib.pyplot as plt
```

```python
# Linear : f(x) = x
# Range x:[-inf, inf] ----> [-inf, inf]
def linear(x):
    return x

# sigmoid: f(x) : 1/(1+e^(-x))
# Range x:[-inf, inf] --> [0, 1]
def sigmoid(x):
    return 1/(1+np.exp(-x))

# tanh: f(x) : (e^x-e^(-x))/(e^x+e^(-x))
# Range x:[-inf, inf] ---> [-1, 1]
def tanh(x):
    return (1-np.exp(-x))/(1+np.exp(-x))

# Relu (Rectified Linear Unit)
# f(x): max(0, x)
# Range x:[-inf, inf] ---> [0, inf]
def relu(x):
    return max(0, x)

# Binary Step Function        (Classification : range[0, 1])--> probability
# f(x): 1 : x>0.5 else 0
# Range x:[-inf, inf] --> 0, 1
def binaryStep(x):
    return x>0.5
```

```python
def plotActivation():
    # Range[-10, 10] ---> 1000 points are even spaces means the differnce betweem the two will
    # always be same.
    x = np.linspace(-10, 10, 1000)
    # fn list storing function in it
    fn = [linear, sigmoid, tanh, relu, binaryStep]
    # fname storing name of the activation function being used
    fname = ['Linear', 'Sigmoid', 'Tanh', 'Relu', 'Binary Step']
    plt.figure(figsize=(10, 10))
    for i in range(5):
        plt.subplot(3, 2, i+1)
        y = np.array([fn[i](element) for element in x])
        plt.title(fname[i])
        plt.plot(x, y)
    plt.show()
```

```python
plotActivation()
```
